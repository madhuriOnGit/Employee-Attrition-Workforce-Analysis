```python
import pandas as pd
import numpy as np
import random

np.random.seed(42)
random.seed(42)
```

```python
departments = [
    'Sales',
    'HR',
    'IT',
    'Finance',
    'Marketing',
    'Operations'
]

job_roles = {
    'Sales': ['Sales Executive', 'Sales Manager'],
    'HR': ['HR Executive', 'Recruiter'],
    'IT': ['Developer', 'Data Analyst', 'System Admin'],
    'Finance': ['Accountant', 'Financial Analyst'],
    'Marketing': ['Marketing Executive', 'SEO Specialist'],
    'Operations': ['Operations Executive', 'Operations Manager']
}
```

```python
data = []

for i in range(1, 5001):

    department = random.choice(departments)

    role = random.choice(job_roles[department])

    age = np.random.randint(22, 60)

    gender = random.choice(['Male', 'Female'])

    years_at_company = np.random.randint(0, 20)

    overtime = random.choice(['Yes', 'No'])

    monthly_income = np.random.randint(25000, 150000)

    job_satisfaction = np.random.randint(1, 6)

    work_life_balance = np.random.randint(1, 6)

    training_hours = np.random.randint(5, 80)

    performance_rating = np.random.randint(1, 6)

    distance_from_home = np.random.randint(1, 40)

    # Attrition Logic
    attrition_score = 0

    if overtime == 'Yes':
        attrition_score += 3

    if job_satisfaction <= 2:
        attrition_score += 3

    if work_life_balance <= 2:
        attrition_score += 2

    if years_at_company < 2:
        attrition_score += 2

    if monthly_income < 40000:
        attrition_score += 2

    attrition_probability = min(attrition_score / 10, 0.9)

    attrition = np.random.choice(
        ['Yes', 'No'],
        p=[attrition_probability, 1 - attrition_probability]
    )

    data.append({
        'employee_id': f'EMP{i:05}',
        'age': age,
        'gender': gender,
        'department': department,
        'job_role': role,
        'monthly_income': monthly_income,
        'years_at_company': years_at_company,
        'overtime': overtime,
        'job_satisfaction': job_satisfaction,
        'work_life_balance': work_life_balance,
        'training_hours': training_hours,
        'performance_rating': performance_rating,
        'distance_from_home': distance_from_home,
        'attrition': attrition
    })

df = pd.DataFrame(data)

df.head()
```

```python
df.shape
```

```python
# Add missing values intentionally

for col in ['monthly_income', 'job_satisfaction', 'work_life_balance']:

    idx = np.random.choice(df.index, 40, replace=False)

    df.loc[idx, col] = np.nan
```

```python
duplicates = df.sample(20)

df = pd.concat([df, duplicates], ignore_index=True)
```

```python
df.to_csv('./data/hr_dataset.csv', index=False)
```
